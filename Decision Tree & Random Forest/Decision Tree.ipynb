```python
#用于数据处理和分析的工具包
import pandas as pd
#引入用于数据预处理/特征工程的工具包
from sklearn import preprocessing
#引入用于数据预处理/特征工程的工具包
from sklearn import tree
```

```python
adult_data = pd.read_csv('DecisionTree.csv')
```

```python
adult_data.head()
```

```python
adult_data.info()
```

```python
adult_data.shape
```

```python
adult_data.columns
```

```python
feature_columns = [u'workclass', u'education', u'marital-status', u'occupation', u'relationship', u'race', u'gender', u'native-country']
# u''转化成unicode编码
label_column = ['income']
```

```python
features = adult_data[feature_columns]
label = adult_data[label_column]
```

```python
features.head()
```

```python
label.head()
```

```python
features = pd.get_dummies(features)
# hot coding
features.head()
```

```python
clf = tree.DecisionTreeClassifier(criterion = 'entropy', max_depth = 4)
# criterion 用熵作为评估准则，最大树的深度为4
clf = clf.fit(features.values,label.values)
# .value 将dataframe格式的features转化为arr的数值型
```

```python
clf
```

```python
clf.predict(features.values)
```

```python
import pydotplus
```

```python
from IPython.display import display, Image
```

```python
dot_data = tree.export_graphviz(clf,
                                out_file=None,
                                feature_names = features.columns,
                                class_names = ['<=50k','>=50k'],
                                filled = True,
                                rounded = True)
```

```python
graph = pydotplus.graph_from_dot_data(dot_data)
display(Image(graph.create_png()))
```
