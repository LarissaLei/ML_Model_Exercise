```python
import pandas as pd
from sklearn import preprocessing
from sklearn.ensemble import RandomForestRegressor
from sklearn.datasets import load_boston
```

```python
boston_house = load_boston()
```

```python
boston_house
```

```python
boston_feature_name = boston_house.feature_names
boston_features = boston_house.data  # 此题的X
boston_target = boston_house.target  # 此题的y
```

```python
boston_feature_name
```

```python
print(boston_house.DESCR)
# DataFrame describe
```

```python
boston_features[:5,:]
```

```python
boston_target
```

```python
rgs = RandomForestRegressor(n_estimators = 15)
# estimator: the number of the tree： 每棵树就是一种样本和采样，都是决策树，准备好了15颗DT的models 
rgs = rgs.fit(boston_features, boston_target)
rgs 
# rgs 训练出来的回归器
```

```python
rgs.predict(boston_features)
```
